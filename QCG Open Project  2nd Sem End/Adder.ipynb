```python
qreg_c0 = QuantumRegister(1, 'c0')
qreg_a1 = QuantumRegister(1, 'a1')
qreg_b1 = QuantumRegister(1, 'b1')
qreg_c1 = QuantumRegister(1, 'c1')
qreg_a2 = QuantumRegister(1, 'a2')
qreg_b2 = QuantumRegister(1, 'b2')
qreg_c2 = QuantumRegister(1, 'c2')
qreg_a3 = QuantumRegister(1, 'a3')
qreg_b3 = QuantumRegister(1, 'b3')
qreg_c3 = QuantumRegister(1, 'c3')
qreg_a4 = QuantumRegister(1, 'a4')
qreg_b4 = QuantumRegister(1, 'b4')
qreg_c4 = QuantumRegister(1, 'c4')

reg_c = ClassicalRegister(5, 'c')

circuit = QuantumCircuit(qreg_c0, qreg_a1, qreg_b1, qreg_c1, qreg_a2, qreg_b2, qreg_c2, qreg_a3, qreg_b3, qreg_c3, qreg_a4, qreg_b4, qreg_c4, reg_c)

circuit.reset(qreg_c0[0])
circuit.reset(qreg_c1[0])
circuit.reset(qreg_c2[0])
circuit.reset(qreg_c3[0])
circuit.reset(qreg_c4[0])

def quantum_full_adder(circuit,c0,a1,b1,c1):
    #This is the complete gate
    circuit.ccx(qreg_c0[0], qreg_a1[0], qreg_c1[0])
    circuit.ccx(qreg_c0[0], qreg_b1[0], qreg_c1[0])
    circuit.x(qreg_c0[0])
    circuit.ccx(qreg_a1[0], qreg_b1[0], qreg_c1[0])
    circuit.ccx(qreg_c0[0], qreg_a1[0], qreg_b1[0])
    circuit.x(qreg_c0[0])
    circuit.x(qreg_a1[0])
    circuit.ccx(qreg_c0[0], qreg_a1[0], qreg_b1[0])
    circuit.x(qreg_a1[0])


#This is the complete gate
circuit.ccx(qreg_c0[0], qreg_a1[0], qreg_c1[0])
circuit.ccx(qreg_c0[0], qreg_b1[0], qreg_c1[0])
circuit.x(qreg_c0[0])
circuit.ccx(qreg_a1[0], qreg_b1[0], qreg_c1[0])
circuit.ccx(qreg_c0[0], qreg_a1[0], qreg_b1[0])
circuit.x(qreg_c0[0])
circuit.x(qreg_a1[0])
circuit.ccx(qreg_c0[0], qreg_a1[0], qreg_b1[0])
circuit.x(qreg_a1[0])

# Now we will add the second gate
circuit.ccx(qreg_c1[0], qreg_a2[0], qreg_c2[0])
circuit.ccx(qreg_c1[0], qreg_b2[0], qreg_c2[0])
circuit.x(qreg_c1[0])
circuit.ccx(qreg_a2[0], qreg_b2[0], qreg_c2[0])
circuit.ccx(qreg_c1[0], qreg_a2[0], qreg_b2[0])
circuit.x(qreg_c1[0])
circuit.x(qreg_a2[0])
circuit.ccx(qreg_c1[0], qreg_a2[0], qreg_b2[0])
circuit.x(qreg_a2[0])

# Now we will add the third gate
circuit.ccx(qreg_c2[0], qreg_a3[0], qreg_c3[0])
circuit.ccx(qreg_c2[0], qreg_b3[0], qreg_c3[0])
circuit.x(qreg_c2[0])
circuit.ccx(qreg_a3[0], qreg_b3[0], qreg_c3[0])
circuit.ccx(qreg_c2[0], qreg_a3[0], qreg_b3[0])
circuit.x(qreg_c2[0])
circuit.x(qreg_a3[0])
circuit.ccx(qreg_c2[0], qreg_a3[0], qreg_b3[0])
circuit.x(qreg_a3[0])

# Now we will add the fourth gate
circuit.ccx(qreg_c3[0], qreg_a4[0], qreg_c4[0])
circuit.ccx(qreg_c3[0], qreg_b4[0], qreg_c4[0])
circuit.x(qreg_c3[0])
circuit.ccx(qreg_a4[0], qreg_b4[0], qreg_c4[0])
circuit.ccx(qreg_c3[0], qreg_a4[0], qreg_b4[0])
circuit.x(qreg_c3[0])
circuit.x(qreg_a4[0])
circuit.ccx(qreg_c3[0], qreg_a4[0], qreg_b4[0])
circuit.x(qreg_a4[0])

#Now we will measure the b bits and c4
circuit.measure(qreg_b1[0], reg_c[0])
circuit.measure(qreg_b2[0], reg_c[1])
circuit.measure(qreg_b3[0], reg_c[2])
circuit.measure(qreg_b4[0], reg_c[3])
circuit.measure(qreg_c4[0], reg_c[4])


circuit.draw()
```
